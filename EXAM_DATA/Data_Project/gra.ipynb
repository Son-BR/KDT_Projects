```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

DF=pd.read_csv('금리 수출입 환율.csv')
DF=DF.T.drop('Unnamed: 0').T
```

```python
for i in DF.iloc[:41,1:]:
    if i=='환율' or i=='미국금리' or i=='한국금리':
        DF[i]=DF[i].astype(float)

    else: DF[i]=DF[i].astype(int)

DF.info()
```

```python
# 미국 수출 비중
ame=DF['대미국수출금액'].mean()
all_ame=DF['수출금액'].mean()-DF['대미국수출금액'].mean()
plt.pie([ame,all_ame], labels=['미국', '그 외 국가'], autopct='%.1f%%')

plt.title('대 미국 수출금액 비중')
plt.show()
```

```python
# 전체 수출 액수 중 미국이 차지하는 비율 연도별
# 전체 subplot의 개수 설정
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('전체 수출 액수 중 미국이 차지하는 비율', fontsize=15)

for i in range(4):
    plt.subplot(2, 2, i+1)
    ame=DF[12*i:12*(i+1)]['대미국수출금액'].mean()
    all=DF[12*i:12*(i+1)]['수출금액'].mean()-DF[12*i:12*(i+1)]['대미국수출금액'].mean()
    plt.pie([ame,all],labels=['미국', '그 외'], autopct='%.1f%%')
```

```python
ame=DF['대미국수출금액'][36:].mean()
all_ame=DF['수출금액'][36:].mean()-DF['대미국수출금액'][36:].mean()
plt.pie([ame,all_ame], labels=['미국', '그 외 국가'], autopct='%.1f%%')

plt.title('대 미국 수출금액 비중')
plt.show()
```

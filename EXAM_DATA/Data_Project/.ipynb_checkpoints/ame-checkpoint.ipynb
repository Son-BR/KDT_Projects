```python
# 파일 불러오기
df=pd.read_excel('./Data/국가별 수출입 실적_202207201.xls',skiprows=4)

df
```

```python
# 총계 지우기
df=df.drop(0).reset_index(drop=True)
# 필요한 열만 추출
df2=df[df.columns[[0,3,5]]]

# 데이터 콤마 제거후 숫자로 변경 
for col in df2.columns[1:]:
    for i in range(df.shape[0]):
        df2[col][i]=df2[col][i].replace(',','')
    df2[col] = pd.to_numeric(df2[col])

# 기간컬럼 형태 변경
for i in range(df2.shape[0]):
    df2['기간'][i]=df2['기간'][i].replace('.','-')

# 컬럼명 변경
df2.rename(columns={'수출금액':'미국수출금액','수입금액':'미국수입금액'},inplace=True)
df2
```

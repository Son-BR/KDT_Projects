```python
import pymysql
import pandas as pd
```

```python
join_table="""select * from user_table as u inner join buy_table as b on u.userID=b.userID"""
```

```python
# 2. 두 테이블을 내부 조인(buy_table.useID와 user_table.userID)한 다음, 아래의 결과와 같이 출력이 되도록 SQL 쿼리를 작성하시오.
conn = pymysql.connect(host='localhost', user='root',password='950603', db = 'shoppingmall', charset='utf8')
cur=conn.cursor()


cur.execute(join_table)

rows = cur.fetchall()
join_df = pd.DataFrame(rows)
print(join_df)

cur.close()
conn.close()
```

```python
join_df=join_df.T.drop_duplicates().T
join_df.drop(8,axis=1,inplace=True)
```

```python
join_df.columns=['userID','userName','birthYear','addr','mobile1', 'mobile2','height','mDate','prodName','groupName','price', 'amount']
join_df
```

```python
# 1) 내부 조인한 결과에 ‘연락처’ 컬럼 추가
join_df['연락처']=join_df['mobile1']+join_df['mobile2']
join_df[['userName','prodName','addr','연락처']]
```

```python
# 2) userID가 KYM인 사람이 구매한 물건과 회원 정보 출력
join_df[join_df['userID']=='KYM'][['userID','userName','prodName','addr','연락처']]
```

```python
# 3) 전체 회원이 구매한 목록을 회원 아이디 순으로 정렬
join_df.sort_values('userID')
```

```python
# 4) 쇼핑몰에서 한 번이라도 구매한 기록이 있는 회원 정보를 회원 아이디 순으로 출력
conn = pymysql.connect(host='localhost', user='root',password='950603', db = 'shoppingmall', charset='utf8')
cur=conn.cursor()

query="""
select distinct u.userID, userName, addr from user_table as u
inner join buy_table as b
on b.userID=u.userID
ORDER BY u.userID;
"""
cur.execute(query)

rows = cur.fetchall()
df = pd.DataFrame(rows)
df.columns=[['userID','userName','addr']]
print(df)

cur.close()
conn.close()
```

```python
# 5) 쇼핑몰 회원 중에서 주소가 경북과 경남인 회원 정보를 회원 아이디 순으로 출력
join_df[(join_df['addr']=='경북')|(join_df['addr']=='경남')].sort_values('userID')
```

```python
import pymysql
import pandas as pd
import matplotlib.pyplot as plt
import platform
```

```python
# 만성 질환 데이터 수정후 저장

conn = pymysql.connect(host='localhost', user='root',password='950603', db = 'project_db', charset='utf8')
cur=conn.cursor()
cur.execute("""select * from disease_sort
where 항목 in ('유병률 (%)','치료율 (%)') and 성별='전체';""")

df=pd.DataFrame(cur.fetchall())
df=df.T
disease=['항목', '성별', '시점', '고혈압', '뇌졸중(중풍 뇌경색)', '고지혈증(이상지질혈증)', '협심증 심근경색증',
       '기타 심장질환', '당뇨병', '갑상선 질환', '골관절염 류머티즘관절염', '골다공증', '요통 좌골신경통',
       '만성기관지염 폐기종(CO포인트D)', '천식', '폐결핵 결핵', '백내장', '녹내장', '만성중이염', '악성신생물(암)',
       '위,십이지장궤양', '간염', '간경변', '만성신장질환', '전립선비대증', '요실금', '성병(매독 등)', '빈혈',
       '피부병', '우울증', '치매', '골절 탈골 및 사고후유증', '불면증', '파킨슨', '노인성 난청', '기타']

df.insert(0,'disease',disease)


df=df.iloc[3:]
df.columns=['disease','유병률2014','유병률2017','유병률2020','치료율2014','치료율2017','치료율2020']

cur.close()
conn.close()

df.replace('-',None,inplace=True)
df.replace('0',None,inplace=True)
df.dropna(inplace=True)
df.to_csv('./data/disease_sort2.csv',encoding='utf-8',index=False)
```

```python
def disease_count(cur):
    # 의사 진단 만성 질환 수
    
    sql="""select 시점,만성질환수별,소계 from disease_doctor
    WHERE 만성질환수별 in ('없음 (%)','1개 (%)','2개 (%)','3개 이상 (%)')
    order by 시점;"""
    
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    fig, axes = plt.subplots(1, 3, figsize=(23, 8))
    fig.suptitle('노인의 만성질병 갯수', fontsize=15)
    year=[2014,2017,2020]
    for i in range(3):
        plt.subplot(1,3,i+1).set_title(year[i])
        plt.pie(df[2][i*4:(i+1)*4],labels=['없음', '1개', '2개', '3개 이상'],autopct='%.1f%%')
    plt.show()
    
def disease_top10(cur):
    # 유병율 상위 10개 만성질병, 치료율
    
    sql="""select disease, 유병률2020, 치료율2020 from disease_sort2
    order by 유병률2020 desc
    limit 10;"""
    
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    plt.figure(dpi=100,figsize=(20,5))
    plt.xticks(rotation=20,fontsize=20)
    plt.title("유병율 상위 10개 만성질병과 치료율",fontsize=20)

    plt.bar(df[0],df[1])
    plt.plot(df[0],df[2])
    plt.legend(['치료율','유병률'],fontsize=10)
    plt.show()
    
def disease_low10(cur):
    # 치료율 하위 10개 만성질병
    
    sql="""select disease, 치료율2020 from disease_sort2
    order by 치료율2020
    limit 10;"""
    
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    plt.figure(dpi=100,figsize=(20,5))
    plt.xticks(rotation=20,fontsize=20)
    plt.title("치료율 하위 10개 만성질병",fontsize=20)

    plt.bar(df[0],df[1])

    plt.show()
    
def blues(cur):
    # 연도별 우울증상
    sql="""select 시점,소계 from blues
    where 증상별 = '우울증상 (%)';"""

    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    plt.figure(dpi=100,figsize=(20,5))
    plt.title("우울증상",fontsize=20)
    plt.plot(df[0],df[1])
    plt.show()

def facility(cur):
    # 의료복지시설의 수와 입소정원
    
    sql="""select 시점, (노인요양시설+노인요양공동생활가정) as 시설수
    from facility
    WHERE 항목='시설수 (개소)';"""
    cur.execute(sql)
    df1=pd.DataFrame(cur.fetchall())
    
    sql2="""select 시점, (노인요양시설+노인요양공동생활가정) as 입소정원
    from facility
    WHERE 항목='입소정원 (명)';"""
    cur.execute(sql2)
    df2=pd.DataFrame(cur.fetchall())
    
    plt.rcParams['figure.figsize'] = (25, 7)

    fig, ax1 = plt.subplots()
    ax1.plot(df1[0],df1[1],color='blue')
    ax1.legend(['시설수'],loc='upper left')
    plt.title('의료복지시설의 수와 입소정원',fontsize=20)

    ax2 = ax1.twinx()
    ax2.plot(df1[0],df2[1],color='red')
    ax2.legend(['입소정원'],loc='upper right')

    plt.show()
    
def treat(cur):
    # 수진율 및 미치료율
    sql="""select 수진율및미치료율별,과목별,소계 from treat_rate
    where 과목별 in ('건강검진','치매검진','병의원 진료','치과 진료') and 시점='2020';"""
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    print(df.head())
    
```

```python
def treat5(cur):
    # 2020 분위별 건강검진
    
    x_label=['제1오분위','제2오분위','제3오분위','제4오분위','제5오분위']
    
    sql="""select 수진율및미치료율별,과목별,제1오분위,제2오분위,제3오분위,제4오분위,제5오분위
    from treat_rate
    where 과목별 = '건강검진' and 시점='2020';"""
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    plt.figure(dpi=100,figsize=(20,5))

    plt.title("2020년 분위별 건강검진 비율",fontsize=20)
    plt.plot(x_label,df.iloc[0][2:])
    plt.show()
    
def disease_count5(cur):
    # 2020 분위별 만성질환 수
    
    x_label=['제1오분위','제2오분위','제3오분위','제4오분위','제5오분위']
    
    sql="""select 제1오분위,제2오분위,제3오분위,제4오분위,제5오분위 from disease_doctor
    WHERE 만성질환수별 = '없음 (%)' and 시점 = '2020';"""
    cur.execute(sql)
    df1=pd.DataFrame(cur.fetchall())
    
    sql="""select 제1오분위,제2오분위,제3오분위,제4오분위,제5오분위 from disease_doctor
    WHERE 만성질환수별 = '3개 이상 (%)' and 시점 = '2020';"""
    cur.execute(sql)
    df2=pd.DataFrame(cur.fetchall())
    
    plt.rcParams['figure.figsize'] = (25, 7)

    fig, ax1 = plt.subplots()
    ax1.plot(x_label,df1.iloc[0],color='blue')
    ax1.legend(['없음'],loc='upper left')
    plt.title('2020년 분위별 만성질환 수',fontsize=20)

    ax2 = ax1.twinx()
    ax2.plot(x_label,df2.iloc[0],color='red')
    ax2.legend(['3개 이상'],loc='upper right')

    plt.show()
    
def blues5(cur):
    # 분위별 우울증상
    
    x_label=['제1오분위','제2오분위','제3오분위','제4오분위','제5오분위']
    
    sql="""select 제1오분위,제2오분위,제3오분위,제4오분위,제5오분위 from blues
    where 증상별='우울증상 (%)' and 시점 = '2020';"""
    cur.execute(sql)
    df=pd.DataFrame(cur.fetchall())
    
    plt.figure(dpi=100,figsize=(20,5))
    plt.title("2020년 분위별 우울증상",fontsize=20)
    plt.plot(x_label,df.iloc[0])

    plt.show()
        
```

```python
def main():
    
    # 한글 글꼴
    if(platform.system() == 'Windows'):
        plt.rc('font', family = 'Malgun Gothic')
    else:
        plt.rc('font', family = 'AppleGothic')
        
    conn = pymysql.connect(host='localhost', user='root',password='950603', db = 'project_db', charset='utf8')
    cur=conn.cursor()
    
    disease_count(cur)
    disease_top10(cur)
    disease_low10(cur)
    blues(cur)
    facility(cur)
    treat(cur)
    
   
    
    cur.close()
    conn.close()    
```

```python
main()
```

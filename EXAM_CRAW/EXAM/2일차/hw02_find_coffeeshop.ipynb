# 과제2: 커피 매장 찾기

```python
# 모듈 불러오기
import pandas as pd

# 파일 불러오기
DF=pd.read_csv('hollys_branches.csv',encoding='utf-8')
DF
```

```python
# 도시 입력받기
city=input('검색할 매장의 도시를 입력하세요:')

# 매장수 count
count=0

# 주소, 번호 저장 리스트
addr_phone=[]

# '주소' 컬럼에 입력받은 도시 있으면 리스트에 저장
for i in range(len(DF)):
    if  city.split(' ')[0] in DF['주소'][i] and city.split(' ')[1] in DF['주소'][i]:
        addr_phone.append(f"[ {count+1}]: ['{DF['주소'][i]}', {DF['전화번호'][i]}]")
        count+=1

# 결과 출력
print(f'검색된 매장 수: {count}')

for i in range(len(addr_phone)):
    print(addr_phone[i])

```

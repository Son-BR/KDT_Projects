```python
# 모듈 로딩
import pandas as pd
import numpy as np

from keras import Sequential              # 모델 객체 
from keras.layers import Dense            # 레이어 객체
from keras.layers import Dropout            # 레이어 객체
from keras.layers import Flatten

from sklearn.model_selection import train_test_split
```

```python
# 데이터 로딩
df=pd.read_csv('../Data/diabetes.csv')
df.info()
```

```python
data=np.array(df.iloc[:,:8])
target=np.array(df['Outcome'])
```

```python
# 검증용 데이터셋 준비
x_train, x_test, y_train, y_test = train_test_split(data, target, stratify=target)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, stratify=y_train)
```

```python
x_train.shape, x_test.shape, x_val.shape
```

```python
# 모델 생성

model=Sequential()
```

```python
# 첫 레이어
model.add(Dense(250, activation='relu', input_shape=(8,)))
```

```python
model.add(Dense(250, activation='relu'))
```

```python
# 출력 레이어
model.add(Dense(2, activation='sigmoid'))
```

```python
model.summary()
```

```python
# 모델 생성
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
# 타겟 라벨 -> 원핫인코딩으로 변환
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit(y_train.reshape(-1,1))
y_train=encoder.transform(y_train.reshape(-1,1))
```

```python
encoder.fit(y_val.reshape(-1,1))
y_val=encoder.transform(y_val.reshape(-1,1))

encoder.fit(y_test.reshape(-1,1))
y_test=encoder.transform(y_test.reshape(-1,1))
```

```python
y_train=y_train.toarray()
y_val=y_val.toarray()
y_test=y_test.toarray()
```

```python
# 학습
model.fit(x_train, y_train, epochs=50,validation_data=(x_val, y_val))
```

```python
test_result=model.evaluate(x_test, y_test)
```

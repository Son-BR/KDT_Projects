```python
from sklearn.datasets import load_iris
```

```python
x, y=load_iris(return_X_y=True)
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
```

```python
# 타겟 라벨 -> 원핫인코딩으로 변환
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit(y.reshape(-1,1))
y=encoder.transform(y.reshape(-1,1))
y=y.toarray()
```

```python
from keras import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
```

```python
## 파라미터로 그리드서치로 확인하고 싶은 값들 추가 
# def get_model(n_hidden, n_neurons, learning_rate,....):

# meta안에 있는 값 사용 가능
def get_model(hidden_layer_dim, n_neurons1, n_neurons2, meta):
    
    # print(type(meta), '\n', meta)
    
    n_features_in_ = meta["n_features_in_"]
    X_shape_ = meta["X_shape_"]
    n_classes_ = meta["n_classes_"]
    
    
    
    model=Sequential()
    model.add(Dense(n_features_in_, activation='relu', input_shape=X_shape_[1:]))
    model.add(Dense(n_neurons1, activation='relu'))
    model.add(Dense(n_neurons2, activation='relu'))
    model.add(Dense(n_classes_, activation='softmax'))
              
    return model
```

```python


clf = KerasClassifier(
    get_model,
    loss="categorical_crossentropy",
    hidden_layer_dim=100,
    n_neurons1=10,
    n_neurons2=10
    # n_neurons....
    # 파라미터 넣기
    )
```

```python
clf.fit(x, y)
```

```python
y_proba = clf.predict_proba(x)
```

```python
y_proba
```

```python
from sklearn.model_selection import GridSearchCV


params = {

    # 딕셔너리형태로 그리드서치할 인수 입력
    # "n_neurons" : [10, 30, 50],
    "n_neurons1" : [30, 50, 100],
    "n_neurons2" : [30, 50, 100],
    "hidden_layer_dim": [50],
    "loss": ["categorical_crossentropy"],
    "optimizer": ["adam"],
    "optimizer__learning_rate": [0.0001, 0.001, 0.1],
}
gs = GridSearchCV(clf, params, refit=False, cv=3, scoring='accuracy')

gs.fit(x, y, epochs=5)
print(gs.best_score_, gs.best_params_)
```

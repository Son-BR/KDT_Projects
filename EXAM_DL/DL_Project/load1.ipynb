```python
from keras.models import load_model
```

```python
import numpy as np
x_test=np.load('x_test.npy')
y_test=np.load('y_test.npy')
```

```python
import matplotlib.pyplot as plt

def test_plt(model, i):
    plt.figure(figsize=(20,5))

    plt.subplot(131)
    plt.imshow(x_test[i].reshape(150,150,3))
    plt.xticks([], [])
    plt.yticks([], [])

    plt.subplot(132)
    pre_out=model.predict(x_test[i].reshape(-1,150,150,3))
    plt.imshow(pre_out.reshape(150,150,3))
    plt.xticks([], [])
    plt.yticks([], [])

    plt.subplot(133)
    plt.imshow(y_test[i])
    plt.xticks([], [])
    plt.yticks([], [])

    plt.tight_layout()
    plt.show()
```

```python
reModel=load_model('model10.h5')
reModel2=load_model('model10r.h5')
```

```python
# 메모리 에러남
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

test_plt(reModel, 100)
```

```python
test_plt(reModel2, 100)
```

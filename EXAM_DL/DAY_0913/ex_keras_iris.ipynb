```python
import tensorflow as tf
from keras import Sequential # 레이어를 선형으로 쌓는 모델 객체
from keras.layers import Dense # 레이어객체
from sklearn.datasets import load_iris
 
import numpy as np
```

```python
# 데이터 로딩
iris=load_iris()
data=iris.data
target=iris.target
print(data.shape, target.shape)
```

```python
# 훈련 데이터셋과 테스트 데이터셋 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, shuffle=True,random_state=42)
```

```python
# 모델 구성
model=Sequential(name='IRIS')

# 각 노드(퍼셉트론, 유닛): 4개 입력 / 1개 출력
# 총 10개가 다음 층의 입력
model.add(Dense(10, activation='relu', input_shape=(4,)))

# 각 노드(퍼셉트론, 유닛): 10개 입력 / 1개 출력
# 총 3개 출력
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
# 모델 생성

# categorical_crossentropy: 원핫인코딩 후 넣어야함
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# 학습
model.fit(x_train, y_train)
```

```python
# 원핫인코딩
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()

y_train2=ohe.fit(y_train.reshape(-1,1)).transform(y_train.reshape(-1,1))
y_train3=y_train2.toarray()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train3, epochs=150)
```

```python
ohe=OneHotEncoder()
y_test2=ohe.fit(y_test.reshape(-1,1)).transform(y_test.reshape(-1,1))
y_test3=y_test2.toarray()

# 예측
print(model.predict(x_test))
print(y_test3)
```

```python
# 평가
model.evaluate(x_test, y_test3)
```

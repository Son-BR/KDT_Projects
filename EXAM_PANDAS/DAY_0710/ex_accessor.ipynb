# 접근자(Accessor) 실습

 - DF/SR에 속성으로 제공되며 다른 객체의 기능/속성을 사용할 수 있도록 해줌
 
 - 종류
   - Series.dt : datetime 속성 사용할 수 있도록 해줌
   - Series.dt : string 관련 메서드 사용할 수 있도록 해줌

```python
# 모듈 로딩
import pandas as pd

# 데이터 생성
# 이름, 생년월일, 나이 3개 컬럼으로 구성
# 10명
# 생년월일 : 동년해, 동일날짜
# 이름 : 영문 (예 : Tom Kim)
```

```python
dtRange=pd.date_range('1995-01-05',periods=10,freq="M")
dtRange
```

```python
ageL=[]
for i in range(10): ageL+=[28]
```

```python
df=pd.DataFrame({'name':['Tom Kim','Gildong Hong','Jerry Park','Woo Choi','Jin KIm','One Park','Twe Park','Three Park','Four Park','Five Park'],
                'birth':dtRange,
                'age':ageL})
```

## str데이터로 구성된 컬럼(Series)의 str 접근자

 - 전체 컬럼의 요소에 str관련 메서드 사용 가능하도록 하는 접근자
 - 사용법
    - 객체변수[컬럼명].str.메서드() => upper, lower, split, replace....
    - 객체변수[컬럼명].str.get(인덱스) => 분리문자리스트에서 특정 요소 추출

```python
df['name'].str.upper(), df['name'].str.lower()
```

```python
df['name'].str.contains('a')
```

### 문자열 분리

```python
df['name'].str.split()
```

```python
nameDF=df['name'].str.split()
```

```python
fn=[]
ln=[]
for i in range(len(nameDF)):
    for j in [0,1]:
        if j==0: fn+=[nameDF[i][j]]
        else:  ln+=[nameDF[i][j]]
            
df['firstn']=fn
df['lastn']=ln
df
```

```python
nameDF=df['name'].str.split()
nameDF[0][0]
```

### 분리된 문자열에서 특정값 가져오기

```python
nameDF.str.get(0), nameDF.str.get(1)
```

```python
df['first_name']=nameDF.str.get(0)
df['last_name']=nameDF.str.get(1)
```

```python
df
```

### 분리된 문자열 DF으로 변환

```python
# str.split()
value=df['name'].str.split(expand=True)
value
```

```python
df2=pd.concat([df,value],axis=1)
df2
```

```python
df2.rename(columns={0:'FirstName',1:'LastName'},inplace=True)
df2
```

## 실습

 - stock-data.csv
 
 - 접근자 사용하여 Data컬럼 분할 (Y, M, D)
 
 - D 기준으로 정렬

```python
stock=pd.read_csv('../Data/stock-data.csv')
stock.info(),
stock.head()
```

```python
stockDate=stock['Date'].str.split('-',expand=True)
stockDate
```

```python
stock2=pd.concat([stock,stockDate],axis=1)
stock2
```

```python
stock2.rename(columns={0:'Y',1:'M',2:'D'},inplace=True)
```

```python
stock2.sort_values('D', inplace=True)
stock2
```

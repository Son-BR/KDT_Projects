## 성별, 등석, 나이에 따른 생존율

### 성별에 따른 생존율

```python
female=(df['sex']=='female').sum() #여자전체
male=(df['sex']=='male').sum() # 남자전체
sexGroup=df.groupby('sex')
femaleSur=(sexGroup.get_group('female')['survived']==1).sum() # 여자생존
maleSur=(sexGroup.get_group('male')['survived']==1).sum() #남자생존
```

```python
print(f'여자생존율 : {round(femaleSur/female*100,2)}%\n남자생존율 : {round(maleSur/male*100,2)}%')
```

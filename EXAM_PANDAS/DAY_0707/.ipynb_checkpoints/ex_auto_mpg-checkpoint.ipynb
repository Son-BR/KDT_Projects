```python
# (2)
mpg['horsepower']=pd.to_numeric(mpg['horsepower'],errors='coerce',downcast='integer')
mpg.info()
```

```python
# horsepower 컬럼 타입 변경 : 연속형 int -> 범주형 category 변환
# 구간 => 라벨 horsepower_low, horsepower_nomal, horsepower_high
mpg['horsepower_category']=pd.cut(x=mpg['horsepower'],
       bins=3,
       labels=['horsepower_low', 'horsepower_nomal', 'horsepower_high'],
       include_lowest=True)
```

### 방법3 : (컬럼의 데이터-평균)/표준편차 ==> Z-score

```python
mpg.dtypes
```

```python
def nor(col_name):
    mpg[col_name +'_nor']=mpg[col_name]/abs(mpg[col_name]).max()
```

```python
for col in mpg.columns:
    if mpg[col].dtype in ['int16','int64','float64']:
        nor(col)
mpg
```

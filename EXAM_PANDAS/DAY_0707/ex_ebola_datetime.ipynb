```python
# 모듈 로딩
import pandas as pd
from datetime import datetime as dt

# 파일 관련 변수 설정
DIR='../Data/'
FILE=DIR+'country_timeseries.csv'
```

```python
# (1) CSV => DataFrame 로딩
ebolaDF=pd.read_csv(FILE)
ebolaDF
```

```python
# (2) 데이터 확인
ebolaDF.info()
```

```python
# (3) 전처리
# (3-1) NaN 데이터 처리
ebolaDF.fillna(0, inplace=True)
```

```python
# (3-2) Data Type 정리
ebolaDF['Date']=pd.to_datetime(ebolaDF['Date'])
ebolaDF.info()
```

```python
# (4) 데이터 가공
# Ebola 바이러스 발병 후 며칠이 지났는지에 대한 컬럼 추가
ebolaDay=pd.date_range('2014-03-22','2015-01-05')
ebolaDay
```

```python
dayList=[]
for i in ebolaDF['Date']:
    dayList+=[list(ebolaDay).index(i)]

ebolaDF['D+']=dayList
ebolaDF
```

```python
ebolaDay2=pd.date_range(ebolaDF['Date'][121],ebolaDF['Date'][0])
dayList=[]
for i in ebolaDF['Date']:
    dayList+=[list(ebolaDay2).index(i)]

ebolaDF['D+']=dayList
ebolaDF
```

```python
dayList2=[]
for i in range(122):
    dayList2+=[datetime.today()-ebolaDF['Date'][i]]
    
ebolaDF['D++']=dayList2
ebolaDF['D++']=ebolaDF['D++'].dt.days
ebolaDF
```

```python
dayList2=[]
for i in range(len(ebolaDF['Date'])):
    dayList2+=[(datetime.today()-ebolaDF['Date'][i])]
    
ebolaDF['D++']=dayList2
ebolaDF['D++']=ebolaDF['D++'].dt.days

ebolaDF
```

```python
ebolaDF['D+++']=(datetime.today()-ebolaDF['Date']).dt.days

ebolaDF
```

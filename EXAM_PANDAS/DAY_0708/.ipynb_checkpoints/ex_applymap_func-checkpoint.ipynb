```python
# 모듈 로딩
import pandas as pd, os
import numpy as np

# 파일 관련 변수
DIR='../Data/'
FILE=DIR+'titanic.csv'

if os.path.exists(FILE):print('File OK')
else: print('File is not Exist')
```

```python
titanic=pd.read_csv(FILE)
titanic.head()
```

```python
subDF=titanic.loc[:5,['Age','Fare']]
subDF['Value']=10
subDF
```

```python
# 사용자 정의 함수
def addValue(x):
    print(f'[DEBUG]----------------\n{x}')
    return x+x

def addTwoValue(x,y):
    print(f'[DEBUG]----------------\n{x},{y}')
    return x+y



def missingValueCheck(sr):
    print(f'[DEBUG]----------------\n{sr}')
    return sr.isnull()

def missingValueSum(sr):
    print(f'[DEBUG]----------------\n{sr}')
    return sr.isnull().sum()

def maxmin(sr)
    print(f'[DEBUG]----------------\n{sr}')
    return sr.max()-sr.min()
```
